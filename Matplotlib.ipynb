```python
import numpy as np
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(y)
```

```python
import matplotlib.pyplot as plt
#sine wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title and labels to the x axis and y axis
plt.plot(x,y)
plt.xlabel("Angle")
plt.ylabel("Sine Value")
plt.title("Sine Wave")
plt.show()
```

```python
#Bar Chart
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['ENGLISH','TAMIL','KOREAN','JAPANESE','CHINESE','THAI']
people=[80,75,50,50,80,60]
ax.bar(languages,people)
plt.show()
```

```python
#pie chart
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['ENGLISH','TAMIL','KOREAN','JAPANESE','CHINESE','THAI']
people=[80,75,50,50,80,60]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

```python
#Scatter Plot
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
# 3D Scatter plot
fig3=plt.figure()
ax=plt.axes(projection='3d')
z= 20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
